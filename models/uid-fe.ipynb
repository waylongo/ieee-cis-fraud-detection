```python
def create_group(df, group, index):
    df["uid_" + str(index)] = 0
    for col in group:
        df["uid_" + str(index)] = df["uid_" + str(index)].astype(str)+'_'+df[col].astype(str)
    return df
```

---

```python
def fe(train_df, test_df):

    print("Begin to fe...")

    uid1_group = [
        'TransactionAmt','ProductCD', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6', 
        'P_emaildomain', 'R_emaildomain', 'C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13','C14',
        'D8','D9','D12','D13','D14','D15','M4','id_01', 'id_03',  'id_04', 'id_06', 'id_09', 'id_10', 'id_11',
        'id_15', 'id_16', 'id_17','id_19','id_20', 'id_31', 'id_35','id_36', 'id_37', 'id_38', 'DeviceType','DeviceInfo'
    ]#c
    
    uid2_group = [
        'TransactionAmt','ProductCD', 'addr1', 'addr2', 'R_emaildomain', 'id_14','id_17', 'id_19', 'id_20',
        'id_30',  'id_31',  'id_32', 'id_33', 'id_34', 'id_35','id_36', 'id_37', 'id_38', 'DeviceType','DeviceInfo'
    ]#s
    
    uid3_group = [
        'ProductCD', 'card2', 'addr1', 'addr2', 'P_emaildomain', 'C1','C2','C3','C4','C5','C6','C7','C8',
        'D4', 'M4','M5','M6'
    ]#W
    
    uid4_group = [
        'ProductCD', 'card2','card3','card4','card5','card6','addr1','addr2'
    ] #W
    
    uid5_group  = [
        'TransactionAmt', 'ProductCD',  'P_emaildomain', 'R_emaildomain', 'C4','C5','C6','C7','C8','C9','C10',
        'C12','C13','C14','D1','D2','D8','D9', 'id_01', 'id_05','id_06','id_09','id_13','id_14','id_17',
        'id_19', 'id_20',  'id_28','id_29',
        'id_30',  'id_31',  'id_32', 'id_33', 'id_34', 'id_35','id_36', 'id_37', 'id_38', 'DeviceType','DeviceInfo'
    ]
   
    
#     uid1_group = [
#         'TransactionAmt','ProductCD', 'addr1', 'addr2', 'P_emaildomain', 'C1', 'C2', 'M4', 'M5',
#         'M6'
#     ]

    
    for df in [train_df, test_df]:
        df = create_group(df, uid1_group, "1")
        df = create_group(df, uid2_group, "2")
        df = create_group(df, uid3_group, "3")
        df = create_group(df, uid4_group, "4")
        df = create_group(df, uid5_group, "5")

    w_uid = ["uid_" + str(i) for i in range(1, 6)]
    for col in tqdm_notebook(w_uid):
        for agg_type in ['count']:
            for agg_col in ['TransactionAmt']:
                new_col_name = col + '_count'
                temp_df = pd.concat(
                    [train_df[[col, agg_col]], test_df[[col, agg_col]]])
                temp_df = temp_df.groupby([col])[agg_col].agg([
                    agg_type
                ]).reset_index().rename(columns={agg_type: new_col_name})

                temp_df.index = list(temp_df[col])
                temp_df = temp_df[new_col_name].to_dict()

                train_df[new_col_name] = train_df[col].map(temp_df)
                test_df[new_col_name] = test_df[col].map(temp_df)


#     use_w_features = ["uid_w_" + str(i) + "_count" for i in range(1, 6)]
#     for df in [train_df, test_df]:
#         df.loc[df.ProductCD!="W", use_w_features] =  0

    return train, test
train, test = fe(train, test)
```

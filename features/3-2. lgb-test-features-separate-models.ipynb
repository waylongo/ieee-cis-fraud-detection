```python
def find_use_cols(train_split, train_y_split, features_columns):
    rm_cols = []
    for col in features_columns:
        if (len(train_split[col].dropna()) < len(train_split) * 0.5):
            rm_cols.append(col)
        if (train_split[col].dtype != 'O'):
            if (np.cov(train_split[col].dropna()) < 0.1):
                rm_cols.append(col)
        if(ks_2samp(train_split[col], train_y_split)[0] < 0.9):
            rm_cols.append(col)
    use_cols = list(set(features_columns) - set(rm_cols))
    return use_cols

POINT_SIZE = 8
def visualize_cols(train_split, train_y_split, features_columns, item):
    for col in features_columns:
        fig = plt.figure()
#         plt.plot(train_split.set_index(['TransactionID'])[col][train_y_split==0], ',b', markersize=5)
#         plt.plot(train_split.set_index(['TransactionID'])[col][train_y_split==1], ',r', markersize=5)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 12], ',b', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 13], ',m', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 14], ',b', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 15], ',m', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 16], ',b', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_split.set_index(['TransactionID'])['DT_month'] == 17], ',m', markersize=POINT_SIZE)
        plt.plot(test_split.set_index(['TransactionID'])[col], ',g', markersize=POINT_SIZE)
        plt.plot(train_split.set_index(['TransactionID'])[col][train_y_split==1], '.r', markersize=5)
        plt.title(item + " - "+ str(col))
#         plt.show()
        fig.savefig("../eda/" + str(item) + "/" + str(item) + "_" + str(col) + ".png")
        plt.cla()
        plt.clf()
        plt.close(fig)
    
    return          
```
